## Task 2.2 

```python
# Dropping the columns

# I have created a variable called columns_Drop and in it I have stored two columns 
columns_Drop = ['Data source', 'Time period']

# With the help of drop() function i have dropped Data source and Time period column as it is not required further

Data = Data.drop(columns=columns_Drop)       # drop() function is imported from pandas
print(Data)
```

```python
# Percentage Data Transformation

# Columns_With_Perc is a variable which holds the list of Columns who have "%" sign

Columns_With_Perc = ['Total', 'Rural (Residence)', 'Urban (Residence)','Poorest (Wealth quintile)', 'Richest (Wealth quintile)']

# Using for loop to travel in the Columns_With_Perc list 
for col in Columns_With_Perc:
    # In this part I am removing the % sign from each entry from the columns in the list
    # astype(float) Function is used to convert it into a numeric value
    Data[col] = Data[col].str.rstrip('%').astype(float)     
    
    
```

```python
# Handling missing values

# Creating a variable named missing_values in which I have stored the sum of null values from each column

missing_values = Data.isnull().sum()    # isnull() is from pandas and it is used to find the null values
print(missing_values) # Checking the null values are present or not

# There are null values in column G,I and J. So I have 2 option either I can drop the rows with null values or fill it

Data.fillna(Data.mean(), inplace = True) # Filling the Null values with mean for that I have used fillna() funtion and this function is from pandas

missing_values = Data.isnull().sum()      # Checking if the null values are dropped
print(missing_values)
```

```python
# Removing duplicate entries

# Dupli_count is a variable which holds the value of sum of duplicates in the data set

Dupli_count = Data.duplicated().sum()

# If there are duplicates then the will enter in the if statement 
if Dupli_count:
    print(f"There are {Dupli_count} duplicate rows in the dataset.") # Printing the duplicate rows from the dataset
    Data.drop_duplicates(inplace=True)                               # Dropping the duplicate rows using drop.duplicates() function 
    print("Duplicated rows have been removed.")
# If there are no duplicates in the dataset then it will enter the else statement
else:
    print("There are no duplicated rows in the dataset.")
```

```python
# Removing Whitespaces

# string_columns is a variable which holds the names of all columns that has text values.
Col_with_String = Data.select_dtypes(include=['object']).columns

# For each column with string values I have remove any leading or trailing
for column in Col_with_String:
    Data[column] = Data[column].str.strip()     # using strip() function to strip down the white spaces
```

```python
# unique_values is a variable which holds unique values from the 'Income Group'column
unique_values = Data['Income Group'].unique()

# Displaying the unique values
unique_values

# In this line I have replaced 'Lowerr middle income (LM)' to 'Lower middle income (LM)' using replace() function
Data['Income Group'] = Data['Income Group'].replace('Lowerr middle income (LM)', 'Lower middle income (LM)')

# Display the updated unique values
unique_values = Data['Income Group'].unique()
unique_values
```

```python
# Replacing spaces in the column names with underscores.
# It was not necessary but it is considered as a good programming practice  
Data.columns = Data.columns.str.replace(' ', '_')
```

```python
# Saving the clean data to a new CSV file named 'clean_A1data.csv' without row indices
Data.to_csv('clean_A1data.csv', index=False)

# Displaying a summary of the 'Data' DataFrame, including column names, data types, and non-null values count.
Data.info()
```

```python
new_df = pd.read_csv("E:\RMIT\PDS\Assignments\Assignments_1\clean_A1data.csv")
new_df.head()
```

```python
# Filtering the columns: 'Income_Group' and 'Total' and storing it into a variable Col_Filter
Col_Filter = new_df[['Income_Group', 'Total']]

# Income_Mean is a variable which hold the values of Mean for each category of Income Group 
Income_Mean = Col_Filter.groupby('Income_Group')['Total'].mean().reset_index()

#Print the Mean values for reference
print(Income_Mean)


# Create a side-by-side boxplot

# Plotting the fig size to width = 11 and height = 7
plt.figure(figsize=(11, 7))


# Using Seaborn to create a boxplot for a visual comparison of 'Total' values across different 'Income_Group'.
# boxplot function is used to plot from seaborn library

sns.boxplot(x='Income_Group', y='Total', data=Col_Filter, order=['Low income (L)', 'Lower middle income (LM)', 'Upper middle income (UM)', 'High income (H)'], palette='Set3')

# Using title function to display the title of the box plot 
plt.title("Percentage of Children with Home Internet Access")

# Giving label to x-axis
plt.xlabel('Income Group')

# Giving label to y-axis
plt.ylabel('Percentage')

# Looping over the mean values for each income category and marking them on the plot
for i, row in Income_Mean.iterrows():
    plt.text(i, row['Total'], f'Mean: {row["Total"]:.2f}', ha='center', va='bottom', fontsize=11)


plt.show()
```

```python


# Calculating the median for Rural and storing into a variable name Rural_M
Rural_M = new_df['Rural_(Residence)'].median()

# Calculating the median for Urban and storing into a variable name Urban_M
Urban_M = new_df['Urban_(Residence)'].median()

# Filtering the data frame to keep only the rows whose Rural Residence value 
# is greater than or equal to the median value Rural_M
Filtered_Rural = new_df[new_df['Rural_(Residence)'] >= Rural_M]

# Sorting the filtered dataframe in descending order according to Rural Residence values
Rural_Sort = Filtered_Rural.sort_values(by='Rural_(Residence)', ascending=False)

# Taking the top ten rows from the sorted dataframe (countries with the highest Rural Residence values).
Top_Rural_C = Rural_Sort.head(10)

# Filtering the data frame to keep only the rows whose Urban Residence value 
# is greater than or equal to the median value Urban_M
Filtered_Urban = new_df[new_df['Urban_(Residence)'] >= Urban_M]

# Sorting the filtered dataframe in descending order according to Urban Residence values
Urban_Sorted = Filtered_Urban.sort_values(by='Urban_(Residence)', ascending=False)

# Taking the top ten rows from the sorted dataframe (countries with the highest Urban Residence values).
Top_Urban = Urban_Sorted.head(10)


print("Top 10 Countries with Highest Percentage for Residence (Rural):")
print(Top_Rural_C[['Countries_and_areas', 'Rural_(Residence)']])


print("\nTop 10 Countries with Highest Percentage for Residence (Urban):")
print(Top_Urban[['Countries_and_areas', 'Urban_(Residence)']])

```

```python


# Filtering the data for Upper middle income (UM) group
Upper_Middle = new_df[new_df['Income_Group'] == 'Upper middle income (UM)']

# Calculating statistics for poor wealth group and storing it into variable named poorest_stats 
poorest_stats = Upper_Middle['Poorest_(Wealth_quintile)'].agg(['mean', 'median', 'std'])

# Calculating statistics for Richer wealth group and storing it into variable named richest_stats 
richest_stats = Upper_Middle['Richest_(Wealth_quintile)'].agg(['mean', 'median', 'std'])


print("Data on Poorest Wealth group in Upper Middle-Income Nations:")
print(poorest_stats)

print("\nData on Richest Wealth Group in Upper Middle-Income Nations:")
print(richest_stats)
```
